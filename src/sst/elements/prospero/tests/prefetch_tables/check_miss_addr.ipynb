```python
f = open('/home/ydy/sst-elements/sst-elements-src/src/sst/elements/prospero/tests/final_txt_results/vgg_512kb_10ns_misscheck.txt', 'r')
# f = open('/home/ydy/sst-elements/sst-elements-src/src/sst/elements/prospero/tests/txt_results_previous/vgg_1mb_500ns.txt', "r")

pc_address = {}

lines = f.readlines()
for line in lines:
    line = list(line.split())
    addr = int(line[3], 16)
    ip = int(line[5], 16)
    if ip in pc_address:
        pc_address[ip].append(addr)
    else:
        pc_address[ip] = [addr]
```

```python
x = ['0-200', '200-400', '400-600', '600-800', '800-1000', '1000-']
y = [0, 0, 0, 0, 0]
vl_x = []
vl_y = []


for hk, hv in pc_address.items():
    if len(hv) >= 10000:
        vl_x.append(hk)
        vl_y.append(len(hv))
    else:
        x.append(hk)
        y.append(len(hv))
```
